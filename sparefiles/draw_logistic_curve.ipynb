```python
const = 5
slope = 1
xvar = np.linspace(0,10,100)

df = pd.DataFrame({'xvar':xvar})

# Step through incrementally steeper slopes (higher coefficients)
for s in range(1, 6):
    df['slope = {}'.format(-s)] = [1/(1+np.exp(-(const + i * -s ))) 
                                   for i in np.linspace(0,10,100)]
    df['slope = {}'.format(s)] = [1/(1+np.exp(-(-const + i * s ))) 
                                   for i in np.linspace(0,10,100)]
df.set_index('xvar', inplace=True)
df.plot.line()
```

```python
const = 5
slope = 1
xvar = np.linspace(0,10,100)

df = pd.DataFrame({'xvar':xvar})

# Step through incrementally intercepts (higher constants)
for c in range(1, 7):
    df['const = {}'.format(-c)] = [1/(1+np.exp(-(c + i * -1 ))) 
                                   for i in np.linspace(0,10,100)]
    df['const = {}'.format(c)] = [1/(1+np.exp(-(-c + i * 1 ))) 
                                   for i in np.linspace(0,10,100)]
df.set_index('xvar', inplace=True)
df.plot.line()
```

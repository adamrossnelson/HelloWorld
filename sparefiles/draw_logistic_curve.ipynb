```python
# Draw logistic curve
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.DataFrame({'xvar':np.linspace(0,6000,1000),
                   'yvar1':[1/(1+np.exp(-(6 + i * -.0025 ))) 
                           for i in np.linspace(0,6000,1000)],
                   'yvar2':[1/(1+np.exp(-(-6 + i * .0025 ))) 
                           for i in np.linspace(0,6000,1000)]})
df.set_index('xvar', inplace=True)
df.plot.line()
```
